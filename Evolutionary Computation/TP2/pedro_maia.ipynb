## Algoritmo *Particle Swarm Optimization*

O algoritmo é inspirado no comportamento social de uma espécie, em que o comportamento de cada indivíduo é baseado tanto em seu próprio conhecimento quanto da população.

O algoritmo é inicializado com indivíduos (chamados de partículas) em posições aleatórias $x_{i,0}$ com aptidões correspondentes $f(x)$. A cada iteração $t$, um vetor de velocidade $v_{i,t}$ é calculado para cada partícula, e a posição é atualizada: $x_{i,t} = x_{i,t-1} + v_{i,t}$.

A forma de cálculo do vetor de velocidades $v_{i,t}$ é fundamental para a boa convergência do algoritmo e é composta por três fatores:
- Inércia: depende da velocidade atual da partícula;
- Memória: depende do melhor ponto conhecido pela partícula;
- Cooperação: depende do melhor ponto conhecido pelos vizinhos da partícula.

Tipicamente duas estruturas de vizinhança são utilizadas no fator de cooperação:
- *Global Best(gbest)*: todas as partículas são consideradas vizinhas da partícula $i$
- *Local Best(lbest)*: apenas as partículas de índice adjacente $i-1$ e $i+1$ são consideradas vizinhas da partícula $i$

Por considerar todo o enxame, a estrutura *gbest* é mais elitista que a *lbest*.

Formalmente, a função genérica do cálculo da velocidade é dada por:

\begin{align*}
    v_{i,t} = \chi[w v_{i,t-1} + c_1 r_1 (p_{i,t} - x_{i,t}) + c_2 r_2 (g_{i,t} - x_{i,t})]
\end{align*}

onde:

- $\chi$: coeficiente de escala da velocidade
- $w$: coeficiente de inércia
- $c_1$: coeficientes de escala da memória
- $c_2$: coeficiente de escala da cooperação
- $r_1, r_2$: compoenentes aleatórios
- $p_{i,t}$: melhor ponto conhecido pelo indivíduo
- $g_{i,t}$: melhor ponto conhecido pela vizinhança de $i$

Ao incluir os três fatores na equação, o algoritmo faz com que as partículas se movam pelo espaço de busca em direção aos melhores pontos conhecidos por ela mesma e pelas vizinhas. A inércia e os componentes aleatórios impedem que essa busca seja unidirecional, e os coeficientes de escala ditam o ritmo da busca. 

Neste trabalho, os seguintes parâmetros serão alterados para avaliar a mudança de desempenho: 
- Estruturas de vizinhança: *gbest* e *lbest*
- Coeficiente de escala: $\chi = 1$ e $0 \leq \chi \leq 1$
- Coeficiente de inércia: $w = 1$ e $0 \leq w \leq 1$

Ao todo, oito diferentes configurações serão testadas em cada problema.

O *PSO* está implementado a seguir, permitindo diversas variações nos parâmetros.

```python
import numpy as np
import numpy.matlib as nm
import pandas as pd
import math
import time
import matplotlib.pyplot as plt

np.random.seed(1000)

class PSO:
    Type = 'global' # determina se o algoritmo é gbest ou lbest
    V0 = 'zero' # velocidade inicial
    Chi = 1 # peso global para a velocidade
    Delta = 1 # coeficiente de velocidade máxima
    W = 1 # peso da inércia
    C1 = 1 # peso da memória
    C2 = 1 # peso da cooperação
    
    def __init__(self, nType, v0, chi, delta, w, c1, c2):
        assert nType in ['global', 'local'], 'Type must be global or local'
        assert v0 in ['zero', 'random'], 'Initial speed must be zero or random'
        
        self.Type = nType
        self.V0 = v0
        self.Chi = chi
        self.Delta = delta
        self.W = w
        self.C1 = c1
        self.C2 = c2
        
    # função para calcular a melhor solução conhecida
    def getg(self, fx, p):
        N = len(fx)
        # atribuição de g para gbest PSO
        if self.Type == 'global':
            return nm.repmat(p[fx == min(fx)],N,1)
        
        # atribuição de g para lbest PSO
        g = np.zeros([np.size(p,axis=0), np.size(p,axis=1)])
        for i in range(N):
            prev = i-1 if i>0 else N-1
            post = i+1 if i<N-1 else 0
            idx = [prev, i, post]
            g[i,:] = p[idx][fx[idx] == min(fx[idx])]
        return g
        
    # algoritmo de otimização
    def optimize(self, fitness, N, xmin, xmax, maxIter):
        n = len(xmin)
        
        # cria o swarm inicial
        x = nm.repmat(xmin,N,1) + nm.repmat((xmax-xmin),N,1)*np.random.rand(N,n)
        fx = np.array([fitness(x[i,:]) for i in range(N)])
        
        fxStar = min(fx)
        xStar = x[fx == fxStar]
        
        # determina a velocidade inicial
        if self.V0 == 'zero':
            v = np.zeros([N,n])
        else:
            v = nm.repmat(xmin,N,1) + nm.repmat((xmax-xmin),N,1)*np.random.rand(N,n)
            
        # calcula velocidade máxima para cada variável
        vmax = self.Delta*(xmax - xmin)
            
        # calcula o p inicial
        p = x.copy()
        fxp = fx.copy()
        
        for k in range(maxIter):
            # avaliação das partículas
            fx = np.array([fitness(x[i,:]) for i in range(N)])
            
            # armazena a melhor solução encontrada
            fxmin = min(fx)
            if fxmin < fxStar:
                fxStar = fxmin
                xStar = x[fx == fxStar]
            
            # atualização de p
            for i in range(N):
                if fxp[i] < fx[i]:
                    p[i,:] = x[i,:]
                    fxp[i] = fx[i]
            
            # cálculo do g
            g = self.getg(fx, p)
            
            for i in range(N):
                r1 = np.random.rand(n)
                r2 = np.random.rand(n)
                
                # calcula a nova velocidade
                v[i,:] = self.Chi*(self.W*v[i,:] + 
                         self.C1*r1*(p[i,:] - x[i,:]) + 
                         self.C2*r2*(g[i,:] - x[i,:]))
                
                # satura a velocidade de acordo com o limite
                v[i,:] = np.minimum(v[i,:], vmax)
                
                # atualiza a posição da partivula
                x[i,:] = x[i,:] + v[i,:]
                
                # realiza reflexão para manter dentro dos limites
                oor = x[i,:] > xmax
                x[i,oor] = xmax[oor]+xmax[oor]-x[i,oor]
                oor = x[i,:] < xmin
                x[i,oor] = xmin[oor]+xmin[oor]-x[i,oor]
                
        # retorna melhor resultado
        fx = np.array([fitness(x[i,:]) for i in range(N)])
        fxmin = min(fx)
        if fxmin < fxStar:
            fxStar = fxmin
            xStar = x[fx == fxStar]
        return (xStar, fxStar)
```

## Definição dos parâmetros do algoritmo

```python
# coeficientes de memória e cooperação
c1 = 0.01
c2 = 0.01

delta = 0.1 # limitante de velocidade
v0 = 'zero' # velocidade inicial das partículas

# estruturas de vizinhança
gbest = 'global'
lbest = 'local'
# coeficientes de escala
chi1 = 1
chi2 = 0.1
# coeficientes de inércia
w1 = 1
w2 = 0.5

popSize = 1000 # número de partículas
maxIter = int(100000/popSize) # número de iterações

dim = 10 # dimensão dos problemas de teste
xmin = -100*np.ones(dim) # valor mínimo das variáveis
xmax = 100*np.ones(dim) # valor máximo das variáveis

nRuns = 31 # número de execuções de cada problema em cada configuração
```

## Definição dos problemas de teste

Os dois problemas de *benchmarking* usados no trabalho foram a função Sphere e a função de Rastrigin.

### Função Sphere

A função Sphere é uma função quadrática unimodal dada por

\begin{align*}
    f(x) = \sum\limits_{i=1}^n x_i^2
\end{align*}

onde $n$ é o número de variáveis do problema.

### Função de Rastrigin

A função de Rastrigin é uma função não convexa multimodal usada para avaliar a performance de algoritmos de otimização. Ela é definida por

\begin{align*}
    f(x) = 10n + \sum\limits_{i=1}^n \left( x_i^2 - 10~cos(2 \pi x_i) \right)
\end{align*}

onde $n$ é o número de variáveis do problema.

As funções estão implementadas a seguir:

```python
def Sphere(x):
    n = len(x)
    return sum(x[i]**2 for i in range(n))

def Rastrigin(x):
    n = len(x)
    return 10*n + sum(x[i]**2 - 10*math.cos(2*math.pi*x[i]) for i in range(n))
```

## Execução dos testes

```python
# define as diferentes configuraçẽos de teste
neightbourhood = [gbest, gbest, gbest, gbest, lbest, lbest, lbest, lbest]
chi = [chi1, chi1, chi2, chi2, chi1, chi1, chi2, chi2]
w = [w1, w2, w1, w2, w1, w2, w1, w2]
nConfigs = 8

# cria os algoritmos com diferentes configurações de parâmetro
pso = []
for i in range(nConfigs):
    pso.append(PSO(neightbourhood[i], v0, chi[i], delta, w[i], c1, c2))
    
meanFxSp = np.zeros(nConfigs)
stdFxSp = np.zeros(nConfigs)
meanFxRa = np.zeros(nConfigs)
stdFxRa = np.zeros(nConfigs)
for i in range(nConfigs):
    fxSp = np.zeros(nRuns)
    fxRa = np.zeros(nRuns)
    for j in range(nRuns):
        _, fxSp[j] = pso[i].optimize(Sphere, popSize, xmin, xmax, maxIter)
        _, fxRa[j] = pso[i].optimize(Rastrigin, popSize, xmin, xmax, maxIter)
    meanFxSp[i] = np.mean(fxSp)
    stdFxSp[i] = np.std(fxSp)
    meanFxRa[i] = np.mean(fxRa)
    stdFxRa[i] = np.std(fxRa)
```

## Resultados obtidos

```python
results = pd.DataFrame(data={'Configuração': ['{}, Chi={}, w={}'.format(neightbourhood[i], chi[i], w[i]) for i in range(nConfigs)], 
                    'Mean Sphere': meanFxSp, 'std Sphere': stdFxSp, 
                     'Mean Rastrigin': meanFxRa, 'std Rastrigin': stdFxRa})
print(results)
```

```python
plt.figure()
plt.plot(meanFxSp, 'x-', label='Sphere')
plt.plot(meanFxRa, 'x-', label='Rastrigin')
plt.legend()
plt.xlabel('Config')
plt.ylabel('Mean fitness')
plt.show()
```

## Análise dos resultados

O algoritmo não foi capaz de encontrar bons resultado para nenhum dos algoritmos, dado que o valor ótimo de ambas as funções é zero. Como esperado, melhores resultados foram obtidos na função unimodal, mais simples de ser resolvida. Além disso, o algoritmo obteve padrões semelhantes de resultado nos dois problemas quando as mesmas configurações eram usadas, o que demonstra consistência.

Os resultados obtidos mostram que para ambas as funções a estrutura global apresentou melhor desempenho. O *gbest* usa mais informações disponíveis que o *lbest*, portanto ele tende a guiar a busca para melhores resultados em menor tempo. Entretanto, dependendo da situação pode ocorrer convergência prematura.

Em relação ao coeficiente de escala de velocidade, melhores resultados foram obtidos quando $\chi = 1$. Um valor elevado para esse parâmetro faz com que as partículas se movam demais no espaço de busca de uma iteração para a seguinte, dificultando a convergência, mas aumentando a exploração. Por outro lado, um valor muito baixo move muito pouco as partículas, o que também dificulta a convergência, mas não auxilia na exploração. Uma boa alternativa para buscar melhores resultados seria na adaptação desse parâmetro ao longo do algoritmo, reduzindo seu valor à medida que o algoritmo converge.

O outro parâmetro alterado foi o coeficiente de inércia $w$, que teve um desempenho dependente do parâmetro $\chi$. Quando $\chi = 1$ e $w = 1$ os melhoers resultados foram alcançados, mas quando $\chi = 0.1$ a variação de $w$ não teve impacto significativo na solução. Quando a inércia é muito elevada, as partículas mudam de direção mais lentamente, atrasando a convergência e aumentando a exploração. Porém, como esse fator é ponderado pelo coeficiente de escala $\chi$ ao final do cálculo de velocidade, seu impacto é menor.
